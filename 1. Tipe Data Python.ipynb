# Set

Set merupakan salah-satu tipe data dalam pemrograman dengan tipe data kolektif atau dapat digunakan untuk menyimpan banyak nilai dalam satu variael.

Ada beberapa hal yang harus diperhatikan dalam membuat tipe data ini diantaranya adalah

1. Nilai yang disimpan unik(jika terdapat nilai yang sama atau duplikat tetap dihitung satu). Jika nilai yang diinputkan ada yang sama maka setelah dimasukan tetap dihitung sebagai satu nilai.
2. Nilai yang disimpan tidak bisa diubah atau not mutable.
3. Bersifat unordered atau tidak berurut sehingga tidak bisa diakses dengan index

**Panduan membuat sebuah set**
Untuk membuat set perintah yang digunakan menggunakan `{}` kurung kurawal.

```python
#membuat set disimpan ke : set1
set1={nilai1,nilai2,nilai3}
```

Sebagai contoh kita akan membuat beberapa set:
1. `set1` berisi nilai : 1,2,3,4
2. `set2` berisi nilai : 1,1,2,3,3,4,4
3. `set3` berisi nilai : 'Aku', 'Bahagia',2,1

````python
# Nomer 1
set1={1,2,3,4}
print(set1)
````

````python
# Nomer 2
set2={1,1,2,3,3,4}
print(set2)
````

````python
# Nomer 3
set3={'Aku','Bahagia',2,1}
print(set3)
````

````python
print(type(set1))
````

Sepert yang sudah dijelaskan diatas bahwa list unordered atau tidak berurut dan tidak memiliki indeks sehingga tidak bisa mengambil suatu nilai yang ada didalam set. Sebagai contoh kita akan mengambil nilai 1 dalam `set1` maka perintah yang digunakan sebagai berikut

````python
print(set1[0])
````

Karena set tidak memiliki index sehingga set juga tidak bisa diubah nilai dalam sebuah set. Sebagai contoh kita akan mengubah nilai 1 di `set1` menjadi 10.

````python
set1[0]=10
````

Dengan ini terbukti bahwa set tidak bisa di ganti nilai dalamnya.

Jika dalam data `list` nilai dalamnya dapat menyimpan list lain. Selanjutnya, akan dicoba apakah set dapat menyimpan nilai dalam berupa set.

````python
set4 ={1,2,3,{2,3,4}}
print(set4)
````

Dengan merujuk pada `error` diatas maka telah di buktikan bahwa set tidak dapat menyimpan nilai dalam yang berupa set juga. Selanjutnya kita akan mencoba membuat `set5` : 1,2,3,4,5 setelah itu akan di tambahkan : 6. Jika sudah ditambahkan maka selanjutnya tambahkan : 7,8. 


*Jelasin fungsi* `add()`,`update()


Oke let's code

````python
set5 ={1,2,3,4,5}
print(set5)
````

Saatnya menambahkan single value yaitu : 6

````python
print('set5 sebelum ditambahkan nilai: ',set5)

#tambahkan nilai 6
set5.add(6)

print('set5 sesudah ditambahkan nilai: ',set5)
````

Menambahkan multiple value : 7,8

````python
print('set5 sebelum ditambahkan nilai: ',set5)

#tambahkan nilai 7,8
set5.update({7,8})

print('set5 sesudah ditambahkan nilai: ',set5)
````

# Sequence Type

## String

String merupakan tipe data sequence. String adalah berisikan karakter untuk mendefinisikan string dapat menggunakan tanda `'' (petik satu)` atau `"" (petik dua)`. Mari lihat kode di bawah ini.

````python
# String

e = 'aku'

print(e,'Tipe datanya adalah: ',type(e))
````

Untuk membuktikan bahwa string adalah tipe data sequence (memiliki indeks) maka akan diberlakukan kode sebagai berikut.

````python
e[0]
````

````python
e[1]
````

Akan tetapi string itu tidak support dengan mengganti elemen didalamnya, hanya bisa untuk melakukan pengambilan data saja.

````python
e[0]='u'
````

Karena string merupakan tipe data squence oleh karena itu dapat di aplikasikan pada perulangan for.

````python
for i in e:
    print(i)
````

## Tuple

Tuple merupakan tipe data kolektif di python. Pada dasarnya Tuple merupakan tipe data yang hampir sama seperti set. Namun beberapa perbedaanya adalah sebagai berikut.
1. Cara membuat tuple dengan menggunakan tanda `()`
2. Tuple memiliki indeks
3. Tuple tidak mutable (nilai dalamnya tidak dapat diganti)
4. Nilai dalamnya dapat memiliki elemen yang sama dan di definisikan sesuatu yang beda dengan indeks yang berbeda juga.

Untuk melihat lebih jelas mari kita coba lihat kode di bawah ini.

````python
# nomer 1 & nomer 4

tuple1=(1,2,3,4,5,1,2,3,'aku','dia')

print(tuple1)
````

````python
print(type(tuple1))
````

Setelah dicek benar bahwa tipe data kolektif tersebut adalah tuple yang didefinisikan menggunakan tanda `()`. Dapat dilihat bahwa tuple dapat memuat elemen dalam yang sama sehingga ditunjukan terdapat elemen `1` yang berulang pada urutan ke 0 dan ke 5.

untuk melihat urutanya dapat mengikuti ilustrasi berikut.

</div>

sehingga kita akan mencoba untuk membuktika pernyataan dua bahwa tuple memiliki indeks dengan kode berikut.

````python
# nomer 2
tuple1[0]
````

````python
tuple1[2]
````

Dari hasil kode di atas menunjukan bahwa tuple memiliki index sehingga dapat dilakukan sebuah nilai didalam tuple itu sendiri. 

Selanjutnya, akan ditunjukan apakah tuple dapat merubah elemen yang ada didalam tuple itu sendiri atau tidak.

````python
# nomer 3

tuple1[0]='satu'

print(tuple1)
````

Dengan ditunjukan dari hasil `error` tersebut bahwa tuple tidak dapat mengganti nilai didalamnya.

**Penggunaan Tuple**

Selain set tuple juga dapat digunakan dalam perulangan for.

````python
tup1=(1,2,3,4,'aku','kamu')

for i in tup1:
  print(i)
````

## List

List merupakan tipe data kolektif di python. List merupakan tipe data kolektif yang paling sering digunakan. Beberpa hal yang harus diperhatikan untuk tipe data ini sebagai berikut.

1. Mendefinisikan tipe data ini menggunakan tanda `[]`
2. Tipe data ini memiliki indeks
3. Dapat diganti elemenya atau Mutable
4. Isi elemen dalamnya dapat sama.

````python
# Nomer 1
list1 = [1,2,3,'aku','dia',1,2,3]

print(list1)
````

````python
print(type(list1))
````

Dapat dilihat bahwa pendefinisian variabel menggunakan tanda `[]` merupakan tipe data kolektif list. Selanjutnya, sama halnya dengan tuple list juga memiliki indeks. mari kita coba sebagai berikut. 

````python
# Nomer 2
list1[0]
````

Dijelaskan bahwa list merupakan tipe data kolektif yang mutable sehingga elemen dalamnya dapat di ubah. Mari kita coba sebagai berikut.

````python
# Nomer 3
print('Sebelum nilai elemen diubah:',list1)
list1[0]='satu'

print('Sesudah nilai elemen diubah:',list1)
````

````python
print('Sebelum nilai elemen diubah:',list1)
list1[1]='dua'

print('Sesudah nilai elemen diubah:',list1)
````

**Penggunaan List**

Selain set dan tuple, list merupakan tipe data kolektif yang sering digunakan dalam perulangan for ini.

## Array

Array merupakan data kolektif dan penting dalam bahasa pemrograman python. Karakteristik dari array dan list hampir sama, dapat diorder(memiliki indeks) setiap elemenya, mutable(elemen dalamnya dapat diganti), dapat menambahkan atau menghapus elemen dalam array. Akan tetapi mereka bukan sesuatu yang sama. Perbedaan yang sangat signifikan adalah array hanya dapat menimpan satu tipe data yang sama. Jika isinya integer maka integer semua. Mari kita rangkum point pada array.

1. Mendefinisikan dengan mengimport library `numpy` dan menggunakan method `np.array([#isi array])`
2. Memiliki indeks
3. Elemen dalamnya dapat diganti (Mutable)
4. Isi elemen dalam dapat sama tapi dalam satu tipe data yang sama.

````python
import numpy as np
````

````python
array1=np.array([1,2,3,1,2,3])
array1
````

````python
array1[0]
````

````python
array1[2]
````

````python
print(array1)

array1[2]=2

print(array1)
````
