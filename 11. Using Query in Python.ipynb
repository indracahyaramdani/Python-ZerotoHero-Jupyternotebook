# Melakukan `Query` SQL

Pada tahap awal kita akan melakukan query dasar pada sql. Untuk tahapanya sebagai berikut.

# Select * (Asterik)/ Semua Kolom

Seperti yang kita tau bahwa * (*asterik*) merupakan operator untuk memanggil seluruh kolom/variabel dalam bahasa pemrograman `SQL`. Sehingga kita akan mencoba untuk melakukan  `query` untuk memanggi semua kolom/variabel yang ada pada tabel df sebagai berikut. 

```python
sqldf("SELECT * FROM df")
```

maka terlihat bahwa semua kolom muncul semua sama seperti pemanggilan df pada pandas.

# Select Beberapa Variabel

Konsep pengambilan beberapa kolom adalah untuk mengurangi data yang tidak perlu di tampilkan. Hal ini juga memperingkas data sehingga mudah dibaca.

Sebagai contoh kita akan mengambil variabel `date` dan `admissions_total`.

```python
sqldf("SELECT date,admissions_total FROM df")
```

# Melakukan Alias pada variabel pada Select

Alias pada variael/kolom berguna untuk mengubah nama variabel dengan nama yang baru. Alias pada variabel/kolom berguna untuk mengurangi kemungkinan penamaan variabel yang bertabrakan atau dapat digunakan untuk menamai variabel yang menggunakan operasi baik itu kali, tambah, atau lainya. Mari kita coba sebagai berikut.

```python
sqldf("SELECT date as tanggal,admissions_total as total_admisi FROM df")
```

Setelah dijalankan maka nama variabel sudah terganti sesuai dengan nama aliasnya.

# Group by

`Group By` merupakan perintah untuk menggabungkan data dengan menggunakan kolom/variabel tertentu, dapat dilakukan satu atau lebih kolom sebagai acuan penggabungan.

Mari kita lihat contoh di bawah ini. 
* Kasusnya kita akan menggabungkan data berdasarkan `state`

```python
sqldf("SELECT state, admissions_total as total_admisi FROM df Group by 1")
```

`SELECT state, admissions_total as total_admisi FROM df Group by 1` group by 1 disini merujuk pada kolom `kesatu` yaitu `state` sehingga dapat diwakili oleh angka `1`. Kadang pada RDBMS lainnya sebagai contoh `PostgreeSQL` harus menyertakan fungsi agregat. Oleh karena itu kita akan mempelajari fungsi agregat di bawah ini.

# Group by with Agregat

`Agregat` merupakan salah satu cara untuk mengambil suatu nilai apakah akan diambil `Sum(Jumlah)`,`Max(Maksimum)`, dsb. Sebagai contoh untuk `query`diatas masih belum benar karena hanya mengambil satu nilai `total_admisi`. Untuk melakukan penjumlahan disemua `state` yang sama maka perlu menambahkan `agregat` **SUM()**. Sehingga dapat diperoleh kode sebagai berikut. 

```python
sqldf("SELECT state, SUM(admissions_total)as total_admisi FROM df Group by 1")
```

Tabel yang dihasilkan tentunya berbeda dengan tabel yang diatasnya lagi karena alasan yang sudah dijelaskan tadi. Mungkin sometimes kamu ingin menambahkan batasan hanya untuk negara tertentu saja sehingga data yang lain tidak di tampilkan untuk kenyamanan dalam mendapatkan hasil yang maksimal. Kita akan bahas penggunaan where untuk melakukan batasan/filtering sebagai berikut.

# Filtering menggunakan Where

Banyak kasus dimana terlalu besar sehingga dapat memperlambat pengambilan informasi pada data karena target analisis hanya pada kota tertentu saja. Sehingga fungsi where ini dapat membantu anda untuk mempermudah pengambilan data. Sebagai contoh kasusnya saya akan mengambil data hanya untuk `state` di `Montana` dan `Utah`. Untuk melakukan filtering ini dapat menggunakan kode di bawah ini.

```python
sqldf("SELECT state, admissions_total as total_admisi FROM df Where state in ('Montana','Utah')")
```

Dengan melihat hasil pada tabel diatas hanya diperoleh data untuk `Utah` dan `Montana`
